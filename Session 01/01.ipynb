```python
import pandas as pd
import numpy as np
```

### Loading Data

```python
data = pd.read_csv("student-mat.csv", sep=";")
```

```python
print(data.head())
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
import matplotlib.pyplot as plt

mean_absences = data['absences'].mean()
std_absences = data['absences'].std()

plt.scatter(data.index, data['absences'], color='blue', alpha=0.6, label='Absences')

# Mean line
plt.axhline(mean_absences, color='red', linestyle='--', label=f'Mean = {mean_absences:.2f}')

# 1 standard deviation lines
plt.axhline(mean_absences + std_absences, color='green', linestyle='--', label=f'+1 Std = {mean_absences + std_absences:.2f}')
plt.axhline(mean_absences - std_absences, color='orange', linestyle='--', label=f'-1 Std = {mean_absences - std_absences:.2f}')

plt.xlabel("Student Index")
plt.ylabel("Number of Absences")
plt.title("Absences Around Mean with 1 Std Deviation")
plt.legend()
plt.show()
```
